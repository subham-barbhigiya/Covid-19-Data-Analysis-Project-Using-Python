# 9a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

```python
sns.histplot(dataframe['gdp_per_capita'], kde=True, bins=30)
plt.title('Distribution of GDP Per Capita')
plt.xlabel('GDP Per Capita')
plt.ylabel('Frequency')
plt.show()
```

# 9b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

```python
sns.scatterplot(data=dataframe, x='gdp_per_capita', y='total_cases', alpha=0.7)
plt.title('total_cases  vs  gdp_per_capita')
plt.xlabel('gdp_per_capita')
plt.ylabel('total_cases')
plt.show()
```

# 9d. Plot a bar plot of 'continent' column with 'total_cases'

```python
sns.catplot(data=dataframe, x='continent', y='total_cases',kind='bar')
plt.title('Total Cases by Continent')
plt.xlabel('continent')
plt.ylabel('total_cases')
plt.xticks(rotation=45)
plt.show()
```
